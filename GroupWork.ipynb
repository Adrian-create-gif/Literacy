```python
import pandas as pd
```

```python
df=pd.read_csv('Literacy.csv')
df
```

```python
df.head()
```

```python
#average literacy level of people per region
df_cleaned = df.iloc[1:].copy() # Remove the first row
df_cleaned['Value'] = pd.to_numeric(df_cleaned['Value'], errors='coerce')
location_value=df_cleaned.groupby('Location')['Value'].mean()
location_value
```

```python
location_value.sort_values(ascending=False)
```

```python
#AVerage National literacy score
df_cleaned['Value'].mean()
```

```python
#Literacy score between women
df_cleaned_women = df_cleaned[df_cleaned['Indicator'].str.contains('Women', na=False)]
df_cleaned_women['Value'].mean()
df_cleaned_women
```

```python
#Literacy score in men
df_cleaned_men = df_cleaned[df_cleaned['Indicator'].str.contains('Men', na=False)]
df_cleaned_men['Value'].mean()
df_cleaned_men
```

```python
#Literacy score for men per region
location_value_men=df_cleaned_men.groupby('Location')['Value'].mean()
location_value_men.sort_values(ascending=False)
```

```python
#Literacy score women per region
location_value_women=(df_cleaned_women.groupby('Location')['Value'].mean())
location_value_women.sort_values(ascending=False)
```

```python
#Difference between men and women literacy score
location_value_men-location_value_women
```

```python
#Comparing literacy scores between men who cannot read at all and those who are literate
df_cleaned_men_cant_read = df_cleaned[df_cleaned['Indicator'].str.contains('Men who cannot read at all ', na=False)]
df_cleaned_men_cant_read['Value']
```

```python
#Literacy score for men who cannot read at all
df_cleaned_men_cannot_read = df_cleaned[df_cleaned['Indicator'].str.contains('Men who cannot read at all', na=False)]
df_cleaned_men_cannot_read['Value'].mean()
```

```python
#Literacy score for men who are literate
df_cleaned_literate_men = df_cleaned[df_cleaned['Indicator'].str.contains('Men who are literate', na=False)]
df_cleaned_literate_men['Value'].mean()
```

```python
#literacy score for women who cannot read
df_cleaned_women_cant_read = df_cleaned[df_cleaned['Indicator'].str.contains('Women who cannot read at all', na=False)]
df_cleaned_women_cant_read['Value'].mean()
```
